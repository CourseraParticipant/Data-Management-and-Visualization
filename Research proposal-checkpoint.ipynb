## Research focus & data set

 In particular, I would like to explore if there is an association between health condition and employment status of the U.S. born citizens. More precisely, I want to know if unemployed persons have more overnight stays in hospital than employed persons. Moreover, I would like to know whether the hospital stays depend not only on the employment status but also on the type of occupation.
 
 KEYWORDS (search terms used): health, sickness absence, hospital visits, (un)employment, occupation

## Literature review

[1] has already explored the association between health issues and employment status. Unlike to my research question, the paper concentrates on the association between the type of employment in terms of duration (e.g. permanent employment vs. temorary employment such as fixed-term, on call, project etc) and specific health issues such as pshycholgocal morbidity and sickness absence. The paper revealed that there is an positive association between temporary employment and psychological morbidity. My research question differs as I am trying to include unemployed subjects into the data as well as I consider occupation type rather than duration of the employment.
[2] goes more into direction of our research question. It explores the association between unemployment rate and use of health services, mortality rate as well as epidemiological issues in Canada. It concludes that based  on the epidemiologic data, there is  a positive association between unemployment and many adverse health outcomes. Overall, the study  shows that mortality, death due to cardiovascular disease and suicide among unemployed men and women occur more often than among either employed people or general population. The study was conducted only in Canada. My data set includes born in USA, so I can explore if the same relationship holds true also for that particular group. Moreover, my question does not take into account death rates. Authors of [3] consider the association from dynamical point of view: for each subjects, they compare her/his health care usage during employment vs. unemployment periods. So the study has more indivudual level than my research question which is more on macro level. Moreover, the study in [3] also refers to data from Canada in period between 1983-89. The findings of the work were that dependeng on the history of the mental treatment the unemployment might have either positive or no association with the usage of health services. In particular, there were no differences in health care use, once subjects with history on some mental health treatment were excluded. There was established a positive association between unemployment and health care use if all subjects (with or without mental desease ) were taken into account for the data analysis.
None of these references considered the data from people born in the USA and association of the occupation type and visits in hospitals.
